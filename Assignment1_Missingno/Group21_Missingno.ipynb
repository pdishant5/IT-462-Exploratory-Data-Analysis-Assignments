## Visualizing Missing Data using different functionalities of missingno

```python
msno.bar(df)
plt.show()
```

# Bar Plot (missingno.bar())
The bar plot offers a clear overview of the missing data in each feature by using vertical bars, where the height represents the number of missing values.


What It Reveals:

Percentage of Missing Data: The length of each bar reflects the percentage or count of missing values in the corresponding feature.
Feature Comparison: It enables easy comparison across features, helping to identify which columns have the highest proportion of missing data.







```python
msno.matrix(df)
plt.show()
```

# Matrix Plot (missingno.matrix())
The matrix plot visualizes missing data in the dataset by displaying it as a matrix, where rows represent individual observations and columns represent features. Missing values are shown as white spaces, while existing data is indicated by shaded areas.

What It Shows:
Vertical Bars (Columns): Each vertical bar corresponds to a feature in the dataset, with its height representing the total number of rows. Thin horizontal lines within each bar represent the rows, allowing you to see missing data at the row level.

White Spaces: Missing data in a feature is depicted by white spaces. The more white space in a column, the more missing data that feature has for its rows.

Shaded Areas (Non-Missing Data): Shaded sections indicate available data. If a column has no white spaces, it means no data is missing for any row in that feature.

First Column on the Left: This column serves as an index, representing each row of data. It helps track missingness across different features.

Bottom Line: The plot may display a heatmap-style line at the bottom to highlight the overall percentage of missing data across all features. Darker areas show features with fewer missing values, while lighter areas indicate features with more missing data.

```python
msno.heatmap(df)
plt.show()
```

# Heatmap (missingno.heatmap())
Description:
The heatmap visualizes the correlation between missing data across different features. It uses a color gradient to indicate the strength of the correlation, with darker colors representing stronger relationships.

What It Reveals:
Correlation of Missing Data: The heatmap highlights how missing data in one feature is related to missing data in other features. When two features often have missing values in the same rows, the correlation between their missingness will appear stronger (darker color).

Dependency Relationships: A high correlation in missing data between features may suggest a dependency or systematic pattern in how the data was gathered, indicating potential connections between those features.

```python
msno.dendrogram(df)
plt.show()
```

# Dendrogram (missingno.dendrogram())
Description:
The dendrogram provides a hierarchical clustering of features based on the similarity of their missing data patterns. It organizes features into a tree-like structure, with the height of each branch representing the degree of similarity between their missing values.

What It Reveals:
Feature Clustering: The dendrogram groups together features that exhibit similar patterns of missing data. Features with missing values in the same rows are clustered closely.

Hierarchical Structure: The branch length in the dendrogram indicates how closely related the missing data patterns are. Shorter branches suggest stronger similarities, while longer branches reflect weaker connections between the features' missing data.
